<font size=3>**Decision Tree**</font>

Decision Trees are a non-parametric supervised learning method used for both classification and regression tasks. The goal is to create a model that predicts the value of a target variable by learning simple decision rules inferred from the data features.

```python
import pandas as pd
from sklearn import tree
import matplotlib.pyplot as plt
% matplotlib inline
```

**Construct decision tree using gini index method**

```python
x= df[features].values
y= df['Outcome'].values
Dsc = tree.DecisionTreeClassifier()
```

**Model Training**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
Dsc= Dsc.fit(x_train,y_train)
```

```python
y_pred= Dsc.predict(x_test)
y_pred
```

```python
#lets predict outcome for new sample
print(Dsc.predict([[0,140,45,35,150,53.6,3.627,35]]))
```

```python
print(Dsc.predict([[1,89,66,23,94,28.1,0.167,21]]))
```

**Evaluation metrics for decision tree classification**

```python
from sklearn.metrics import classification_report,confusion_matrix
from sklearn import metrics
print(classification_report(y_test,y_pred))
print('Accuracy is:',metrics.accuracy_score(y_test, y_pred))
```

```python
cnf=confusion_matrix(y_test,y_pred)
cnf
```

```python
#construct Heatmap for confusion matrix
fig,ax=plt.subplots()
sns.heatmap(cnf,annot=True,fmt='g')
ax.set_xlabel('Predicted')
ax.set_ylabel('Actual')
ax.set_title('Confusion Matrix',y=1.1)
plt.tight_layout()
```

```python
#find out probabilities for outcome
y_proba = Dsc.predict_proba(x_test)[:,1]
```

```python
# lets find out probabilities of outcome for ne samples
Dsc.predict_proba([[1,89,66,23,94,28.1,0.167,21]])[:,1]
```

```python
Dsc.predict_proba([[2,100,150,23,150,46.1,0.167,46]])[:,1]
```

**Receiver operating characteristic**

```python
fpr,tpr,_ = metrics.roc_curve(y_test,y_proba)
auc = metrics.roc_auc_score(y_test,y_proba)
plt.plot([0,1],[0,1],'r--')
plt.plot(fpr,tpr,label='data1,auc='+str(auc))
plt.legend(loc=4)
```

```python
print('Accuracy of DS=',metrics.accuracy_score(y_test,y_proba))
```

To read this decision tree, each condition branches left for "true" and right for "false". When you end up at a value, the value array represents how many samples exist in each target value. So value = [0. 5.] mean there are 0 "no diabes" and 5 "diabetes" by the tim we get to that point. value = [3. 0.] means 3 no-diabetes and 0 diabetes.

**Draw Decision tree using GraphViz**

```python
from IPython.display import Image
from sklearn.externals.six import StringIO
import pydotplus

dot_data = StringIO()
tree.export_graphviz(Dsc, out_file=dot_data,feature_names = features)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

**Construct Decision Tree using Entropy**

```python
#model training
Dsc1 = tree.DecisionTreeClassifier(criterion='entropy')
Dsc1.fit(x_train,y_train)
```

```python
#predicting Outcome
y_pred1= Dsc1.predict(x_test)
y_pred1
```

```python
#predicting Outcome for new sample 
print(Dsc1.predict([[0,140,45,35,150,53.6,3.627,35]]))
```

**Evaluation metrics for decision tree using Entropy**

```python
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
print('Accuracy is:',metrics.accuracy_score(y_test, y_pred1))
```

Accuracy of DS using entropy is greater than gini index for this particular case

## Ensemble learning: using a random forest
We'll use a random forest of 10 decision trees to predict employment of specific candidate profiles:

```python
from sklearn.ensemble import RandomForestClassifier
```

**Model Training of Random Forest**

```python
rfc = RandomForestClassifier(n_estimators=100, bootstrap = True)
rfc.fit(x_train,y_train)
```

```python
df.head()
```

```python
#predicting outcome for new samples
print(rfc.predict([[6,148,72,35,0,33.6,0.627,50]]))
```

```python
print(rfc.predict([[0,140,45,35,150,53.6,3.627,35]]))
```

```python
print(rfc.predict([[1,89,66,23,94,28.1,0.167,21]]))
```

```python
y_proba2 = rfc.predict_proba(x_test)[:,1]
```

**Evaaluation Metrics**

```python
y_pred2= rfc.predict(x_test)
y_pred2
cnf=confusion_matrix(y_test,y_pred2)
print(cnf)
print('Accuracy is:',metrics.accuracy_score(y_test, y_pred2))
```

Accuracy of Random forest algorithm is greater than decision tree algorithm. That is why Random forest is used when it comes to accuracy and better performanace

```python
sns.heatmap(cnf,annot=True,fmt='g')
```

```python
print(classification_report(y_test,y_pred2))
```

```python
fpr,tpr,_ = metrics.roc_curve(y_test,y_proba2)
auc = metrics.roc_auc_score(y_test,y_proba2)
plt.plot([0,1],[0,1],'r--')
plt.plot(fpr,tpr,label='data1,auc='+str(auc))
plt.legend(loc=4)
```

The final testing ROC AUC for the random forest was 0.84 compared to 0.70 for the single decision tree with an unlimited max depth. 
Although the random forest overfits (doing better on the training data than on the testing data), it is able to generalize much better to the testing data than the single decision tree. The random forest has lower variance (good) while maintaining the same low bias (also good) of a decision tree.
